```python
def get_car_spec(comb):

    spec = ["attribute(task,root,value).", "attribute(number_rows,root,406)."]

    for i in range(0, len(comb)):
        spec += car_attrs[comb[i]]

    spec += ['entity(view,root,v0).', 'entity(mark,v0,m).']
    
    for i in range(0, len(comb)):
        spec += ['entity(encoding,m,e'+str(i)+').'] + ['attribute((encoding,field),e'+str(i)+',' + car_attr_to_name[comb[i]] + ').']

    spec += ['#show entity/3.', '#show attribute/3.']

    spec = "\n".join(spec)

    return spec
```

```python
def get_weather_spec(comb):

    spec = ["attribute(task,root,value).", "attribute(number_rows,root,1461)."]

    for i in range(0, len(comb)):
        spec += weather_attrs[comb[i]]

    spec += ['entity(view,root,v0).', 'entity(mark,v0,m).']
    
    for i in range(0, len(comb)):
        spec += ['entity(encoding,m,e'+str(i)+').'] + ['attribute((encoding,field),e'+str(i)+',' + weather_attr_to_name[comb[i]] + ').']

    spec += ['#show entity/3.', '#show attribute/3.']

    spec = "\n".join(spec)

    return spec
```

```python
def get_movie_spec(comb):

    spec = ["attribute(task,root,value).", "attribute(number_rows,root,3201)."]

    for i in range(0, len(comb)):
        spec += movie_attrs[comb[i]]

    spec += ['entity(view,root,v0).', 'entity(mark,v0,m).']
    
    for i in range(0, len(comb)):
        spec += ['entity(encoding,m,e'+str(i)+').'] + ['attribute((encoding,field),e'+str(i)+',' + movie_attr_to_name[comb[i]] + ').']

    spec += ['#show entity/3.', '#show attribute/3.']

    spec = "\n".join(spec)

    return spec
```

```python
weight_files = [txt_file for txt_file in os.listdir('./baseline-plus-one-train-weight-outputs/') if txt_file.endswith('.txt')]
weights = {}

for weight_file in weight_files:
    file_name = weight_file.split('.')[0]

    if file_name.startswith("plus"):
        paper = file_name.split('.')[0].split('-')[1]
    else:
        paper = file_name

    weights[paper] = {}

    fr = open('./baseline-plus-one-train-weight-outputs/' + weight_file, 'r')
    lines = fr.readlines()

    for line in lines:
        weight_name = line.split(" ")[1]
        weight_value = int(line.split('=')[1].strip().strip('.'))
        weights[paper][weight_name] = weight_value
```

```python
poss = car_poss + weather_poss + movie_poss

print (poss[0:85])
print (poss[85:110])
print (poss[110:])
```

```python
baseline_plus_one_rec_rank_correlation = {}
for paper in weights:
    if paper == "baseline-APT":
        continue
    baseline_plus_one_rec_rank_correlation[paper] = []

for i in range(0, len(poss)):
    ## get default rec from draco
    comb = poss[i]
    if i < 85:
        input_spec = get_car_spec(comb)
    elif i < 110:
        if i == 85:
            print (comb)
        input_spec = get_weather_spec(comb)
    else:
        if i == 110:
            print (comb)
        input_spec = get_movie_spec(comb)
    
    default_rec = {}
    default_rec_score = []
    default_rec_pref = {}
    i = 1
    for model in d.complete_spec(input_spec, 100):
        vis = dict_to_facts(answer_set_to_dict(model.answer_set))
        default_rec[i] = vis
        default_rec_score.append(model.cost[0])
        default_rec_pref[i] = d.count_preferences(vis)
        i += 1
        
    too_many_cooks_scores = {}
    for paper in weights:
        temp = []
        for rank in default_rec_pref:
            score = 0
            for soft in default_rec_pref[rank]:
                score += weights[paper][soft+"_weight"] * default_rec_pref[rank][soft]
            temp.append(score)
        too_many_cooks_scores[paper] = temp
        
    baseline_rank = ss.rankdata(too_many_cooks_scores["baseline-APT"])
    for paper in too_many_cooks_scores:
        if paper == "baseline-APT":
            continue
        plus_one_rank = ss.rankdata(too_many_cooks_scores[paper])
        res = ss.kendalltau(baseline_rank, plus_one_rank)
        if math.isnan(res.correlation):
            baseline_plus_one_rec_rank_correlation[paper].append(0)
        else:
            baseline_plus_one_rec_rank_correlation[paper].append(abs(res.correlation))
        
print (baseline_plus_one_rec_rank_correlation)
```

```python
plus_one_correlation = {}

for paper in baseline_plus_one_rec_rank_correlation:
    plus_one_correlation[paper] = sum(baseline_plus_one_rec_rank_correlation[paper]) / len(baseline_plus_one_rec_rank_correlation[paper])
    
rec_shift_order = sorted(plus_one_correlation.keys(), key=plus_one_correlation.get)
print (rec_shift_order)
print ({k: v for k, v in sorted(plus_one_correlation.items(), key=lambda item:item[1])})
```

```python
fig, ax = plt.subplots()
data_to_plot = []

for paper in rec_shift_order:
    data_to_plot.append(baseline_plus_one_rec_rank_correlation[paper])

ax.violinplot(data_to_plot)
x_positions = list(range(len(rec_shift_order)))
ax.xaxis.set_ticks(x_positions)
ax.xaxis.set_ticklabels(rec_shift_order)
plt.gca().xaxis.tick_bottom()
plt.xticks(rotation=270,fontsize=40)
plt.yticks(fontsize=40)

fig.set_size_inches(40, 30)
fig.show()

plt.show()
```
