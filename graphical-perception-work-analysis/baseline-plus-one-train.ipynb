```python
import json
import pandas as pd
from typing import Dict, List, Any
import os

from draco import Draco

import numpy as np
from sklearn import svm

import plotly.figure_factory as ff
import matplotlib.pyplot as plt
import matplotlib.colors as colors

from draco.data_utils import pairs_to_vec
```

```python
default_draco = Draco()
```

```python
def train_model(data: pd.DataFrame, test_size: float = 0.3, C: float = 1, quiet=False):

    X_train = data.negative - data.positive

    size = len(X_train)

    y_train = np.ones(size)

    # flip a few examples at random
    idx = np.ones(size, dtype=bool)
    idx[: int(size / 2)] = False
    np.random.shuffle(idx)

    X_train[idx] = -X_train[idx]
    y_train[idx] = -y_train[idx]

    clf = svm.LinearSVC(C=C, fit_intercept=False)
    clf.fit(X_train, y_train)

    if not quiet:
        print("Train score: ", clf.score(X_train, y_train))
        if test_size > 0:
            print("Dev score: ", clf.score(X_dev, np.ones(len(X_dev))))

    return clf
```

```python
## baseline - APT

baseline = {}

with open("../draco2-training-datasets/Mackinlay1986automating(T).json") as file:
    i = 0
    json_data = json.load(file)
    
    for pair in json_data:
        pair["source"] = "Mackinlay1986automating"
        pair["pair_id"] = f'{pair["source"]}_{i}'
        baseline[pair["pair_id"]] = pair
        i += 1

baseline_train_data = pairs_to_vec(baseline)

assert set(baseline_train_data.negative.columns) == set(
    default_draco.soft_constraint_names
), "Feature names do not match."

clf = train_model(baseline_train_data, test_size=0)

features = baseline_train_data.negative.columns
baseline_weights = {}

with open('./baseline-plus-one-train-weight-outputs/baseline-APT.txt', 'w') as fout:
    for feature, weight in zip(features, clf.coef_[0]):
        baseline_weights[f"{feature}_weight"] = int(weight * 1000)
        fout.write(f"#const {feature}_weight = {int(weight * 1000)}.\n")
```

```python
## plus one models

json_files = [pos_json for pos_json in os.listdir('../draco2-training-datasets/') if pos_json.endswith('.json')]

num_files = len(json_files)

for i in range(0, num_files):

    json_file = json_files[i]

    if json_file == "Mackinlay1986automating(T).json":
        continue

    fr = open('../draco2-training-datasets/' + json_file, 'r')
    fdata = json.load(fr)

    plus_one = {}
    pair_id = 0
    
    for pair in fdata:
        pair["source"] = json_file.split('.')[0]
        pair["pair_id"] = f'{pair["source"]}_{pair_id}'
        plus_one[pair["pair_id"]] = pair
        pair_id += 1
        
    combine_data = baseline | plus_one

    train_data = pairs_to_vec(combine_data)

    assert set(train_data.negative.columns) == set(
        default_draco.soft_constraint_names
    ), "Feature names do not match."

    clf = train_model(train_data, test_size=0)

    features = train_data.negative.columns

    with open('./baseline-plus-one-train-weight-outputs/plus-' + json_file.split('.')[0] + '.txt', 'w') as fout:
        for feature, weight in zip(features, clf.coef_[0]):
            fout.write(f"#const {feature}_weight = {int(weight * 1000)}.\n")
```

```python
## train with all papers

json_files = [pos_json for pos_json in os.listdir('../draco2-training-datasets/') if pos_json.endswith('.json')]

num_files = len(json_files)

all_papers = {}

for i in range(0, num_files):

    json_file = json_files[i]

    if json_file == "Mackinlay1986automating(T).json":
        continue

    fr = open('../draco2-training-datasets/' + json_file, 'r')
    fdata = json.load(fr)
    
    pair_id = 0
    
    for pair in fdata:
        pair["source"] = json_file.split('.')[0]
        pair["pair_id"] = f'{pair["source"]}_{pair_id}'
        all_papers[pair["pair_id"]] = pair
        pair_id += 1
        
combine_data = baseline | all_papers

train_data = pairs_to_vec(combine_data)

assert set(train_data.negative.columns) == set(
    default_draco.soft_constraint_names
), "Feature names do not match."

clf = train_model(train_data, test_size=0)

features = train_data.negative.columns

with open('all-papers-weights.txt', 'w') as fout:
    for feature, weight in zip(features, clf.coef_[0]):
        fout.write(f"#const {feature}_weight = {int(weight * 1000)}.\n")
```

```python
## get raw weight

    
fr = open('all-papers-weights.txt', 'r')
lines = fr.readlines()

all_papers_weights = {}

for line in lines:
    weight_name = line.split(" ")[1]
    weight_split = weight_name.split("_")[:-1]
    wname = "_".join(weight_split)
    weight_value = int(line.split('=')[1].strip().strip('.'))
    all_papers_weights[wname] = weight_value
    
fr = open('draco-default-weights.txt', 'r')
lines = fr.readlines()

draco_default_weights = {}

for line in lines:
    weight_name = line.split(" ")[1]
    weight_split = weight_name.split("_")[:-1]
    wname = "_".join(weight_split)
    weight_value = int(line.split('=')[1].strip().strip('.'))
    draco_default_weights[wname] = weight_value
    
all_papers_weights_trim = {}
draco_default_weights_trim = {}

for wname in all_papers_weights:
    if all_papers_weights[wname] == 0 and draco_default_weights[wname] == 0:
        continue
    else:
        all_papers_weights_trim[wname] = all_papers_weights[wname]
        draco_default_weights_trim[wname] = draco_default_weights[wname]
```

```python
titles = ['all-papers', 'draco-default']
soft_arr = all_papers_weights_trim.keys()
heatmap = []

w = []
for soft in soft_arr:
    if all_papers_weights_trim[soft] > 0:
        w.append(1)
    elif all_papers_weights_trim[soft] < 0:
        w.append(-1)
    else:
        w.append(0)
heatmap.append(w)

w = []
for soft in soft_arr:
    if draco_default_weights_trim[soft] > 0:
        w.append(1)
    elif draco_default_weights_trim[soft] < 0:
        w.append(-1)
    else:
        w.append(0)
heatmap.append(w)


figg, ax = plt.subplots()
cax = ax.matshow(heatmap, norm=colors.Normalize(vmin=-2, vmax=2), cmap=plt.cm.RdBu_r)

x_positions = list(range(len(soft_arr)))
ax.xaxis.set_ticks(x_positions)
ax.xaxis.set_ticklabels(soft_arr)
plt.gca().xaxis.tick_bottom()
plt.xticks(rotation=270,fontsize=40)

y_positions = list(range(len(titles)))
ax.yaxis.set_ticks(y_positions)
ax.yaxis.set_ticklabels(titles)
plt.yticks(fontsize=40)

figg.set_size_inches(80, 25)
figg.show()

plt.savefig('visualization-preference-all-papers-vs-draco-default.png')
```
